```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(df["year"], df["per capita income (US$)"], color="red", marker = "+")
plt.show()
```

```python
#Reshape the input features (year) into a 2D array
X= df[["year"]] #using double brackets to select a Dataframe

#Target variable(per capita income (US$))
y = df["per capita income (US$)"] 

#Create a linear regression object
reg = linear_model.LinearRegression()

#Fit the linear regresion model
reg.fit(X, y)

#Plotting
plt.scatter(X, y, color="blue") #Scatter plot of the data points
plt.plot(X, reg.predict(X), color="red", linewidth=2)
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Linear Regression")
plt.show()
```
